```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read data from CSV file
data1 = pd.read_csv('fundamental_diagram_single_lane.csv')
data2= pd.read_csv('fundamental_diagram_single_lane1.csv')
data3 = pd.read_csv('fundamental_diagram_single_lane2.csv')
data4 = pd.read_csv('fundamental_diagram_single_lane3.csv')
data5= pd.read_csv('fundamental_diagram_single_lane4.csv')
data6 = pd.read_csv('fundamental_diagram_single_lane5.csv')
data7 = pd.read_csv('fundamental_diagram_single_lane6.csv')
data8= pd.read_csv('fundamental_diagram_single_lane7.csv')
data9 = pd.read_csv('fundamental_diagram_single_lane8.csv')
data10 = pd.read_csv('fundamental_diagram_single_lane9.csv')

# Extract wavelengths and absorbance values from the CSV data
density1 = data1['Density']
throughput1 = data1['Throughput']
density2 = data2['Density']
throughput2 = data2['Throughput']
density3 = data3['Density']
throughput3 = data3['Throughput']
density4 = data4['Density']
throughput4 = data4['Throughput']
density5 = data5['Density']
throughput5 = data5['Throughput']
density6 = data6['Density']
throughput6 = data6['Throughput']
density7 = data7['Density']
throughput7 = data7['Throughput']
density8 = data8['Density']
throughput8 = data8['Throughput']
density9 = data9['Density']
throughput9 = data9['Throughput']
density10 = data10['Density']
throughput10 = data10['Throughput']

#Average values

average_density = (density1+density2+density3+density4+density5+density6+density7+density8+density9+density10)/10
average_throughput = (throughput1+throughput2+throughput3+throughput4+throughput5+throughput6+throughput7+throughput8+throughput9+throughput10)/10

# Create figure and axis objects
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the absorbance, transmittance, and fluorescence spectrum
ax.plot(average_density, average_throughput, color='darkgreen')

ax.set_title('Fundamental Diagram', size=20)
ax.set_xlabel('Density')
ax.set_ylabel('Throughput')
ax.grid(True)

plt.savefig("Fundamental Diagram")
plt.tight_layout()
plt.show()
```
