```python
import numpy as np
import matplotlib.pyplot as plt

# Example stack of arrays: (time, position, lanes)
time_steps = 100
positions = 50
lanes = 2
data = np.random.randint(0, 2, (time_steps, positions, lanes))

# Plot for each lane
fig, axs = plt.subplots(lanes, 1, figsize=(10, 8), sharex=True)

for lane in range(lanes):
    lane_data = data[:, :, lane]  # Extract data for this lane
    im = axs[lane].imshow(
        lane_data.T, 
        aspect='auto', 
        cmap='gray', 
        origin='lower'
    )
    axs[lane].set_title(f"Lane {lane + 1}")
    axs[lane].set_ylabel("Position")
    fig.colorbar(im, ax=axs[lane], orientation='vertical')

axs[-1].set_xlabel("Time Steps")
plt.tight_layout()
plt.show()
```
